## **`XGBoost` 모형 적합**

&nbsp; 설명변수들이 전부 수치형 변수이므로, `XGBoost`를 이용하여 모형 적합하는 것의 예측 성능이 가장 뛰어날 것으로 판단된다. 부스팅 알고리즘은 변수 간 비선형 관계도 잘 설명할 수 있으므로, 표준화 등 변수의 변환은 수행하지 않고 원 데이터 그대로 분석하도록 하겠다.

### **1. imports**

```python
import pandas as pd
import numpy as np
import warnings

import matplotlib.pyplot as plt
import sklearn.linear_model
from sklearn.model_selection import train_test_split, ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

warnings.filterwarnings("ignore", category=UserWarning)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
HouseSales = pd.read_csv("/root/ML2024/hw/kc_house_data.csv")
HouseSales = HouseSales.drop("price", axis = 1).assign(price = HouseSales.price)
```

* `date` 열은 분석을 위해 다음과 같은 문자형 자료로 변환하였다. 분석 시 해당 열은 `pandas` 모듈의 `get_dummies`함수를 사용하여 더미변수로 처리하였다. 이 경우, `drop_first = True`로 처리하여 기준이 되는 범주를 1월로 지정하였다.

* 또한 `id`열은 단순히 주택 판매 건에 대한 고유 번호로서 모형에 추가하지 않는 것이 바람직하다. 하지만, 해당 보고서에서는 Shrinkage method가 불필요한 설명변수를 얼마나 잘 제거하는지 알아보기 위하여 별도로 제거하지 않았다.

&nbsp; 그리하여 모든 변수가 수치형으로 처리된 데이터프레임을 얻을 수 있었다.

```python
month_cols = pd.get_dummies(pd.to_datetime(HouseSales.date).dt.month,
                            drop_first = True).astype(int)
df_processed = pd.concat([HouseSales.drop("date", axis = 1), month_cols], axis = 1)\
.drop("price", axis = 1).assign(price = HouseSales.price)

print(df_processed.head(10))
```

### **문제 C**

* **선정된 데이터를 $7:3$으로 쪼개서 $7$은 훈련 데이터로, $3$은 평가 데이터로 해서 $\text{RIDGE}$와 $\text{LASSO}$를 시행하시오.**

---

&nbsp; 해당 데이터는 `sklearn`의 `train_test_split` 함수를 사용하여 훈련데이터를 더 크게 하는 7:3으로 분할되었다. 이후 분석에서 동일한 훈련-평가 데이터를 활용하기 위해 매개변수로 `random_state = 14107`을 지정하였으며, 설명변수들과 예측변수는 훈련/평가 데이터에서 각각 `numpy` 모듈의 행렬 객체로 변환하여 `(X, y)`, `(XX, yy)`로 할당하였다.

&nbsp; 또한, 더미 변수를 제외한 수치형 설명변수들은 train set을 기준으로 표준화를 진행하였다. 그리고 test set의 수치형 설명변수를 변환할 때에도 train set에서 적합 후 사용한 표준화 장치를 그대로 적용하였다.

&nbsp; 예측변수의 경우에는 해석을 용이하게 하기 위해 표준화하지 않았다.

```python
## train/test set split
df_train, df_test = train_test_split(df_processed, test_size = 0.3,
                                     shuffle = True, random_state = 14107)
features_list = list(df_train.columns)[:-1] ## save features name

## fitting scaler with train set
numeric_features = df_train.loc[:, :"sqft_lot15"]
scalr = StandardScaler(with_mean = True, with_std = True)
standarized_features = scalr.fit_transform(numeric_features)

## train data
X = np.concatenate([standarized_features, 
                    np.asarray(df_train.loc[:, 2:12])], axis = 1) ## features
y = np.asarray(df_train.price) ## target

## test data
XX = np.concatenate([scalr.transform(df_test.loc[:, :"sqft_lot15"]),
                     np.asarray(df_test.loc[:, 2:12])], axis = 1)
yy = np.asarray(df_test.price)
```

#### **문제 1**

1. **훈련 데이터에서 $10\text{-fold CV}$를 이용하여 축소 정도를 결정하는 최적의 $\lambda$를 $\text{CV plot}$을 그려서 $1\text{-se}$ 규칙으로 선택하시오. (코드 및 결과)**

---

&nbsp; 10-fold CV를 이용하여 CV error를 최소화하는 $\lambda$ 값을 찾은 후, $1\text{-se}$ rule로 최적의 $\lambda$를 선택하였다.

`-` **Ridge**

```python
## fitting and cross validation(10-fold)
lambdas = 10**np.linspace(6, -4, 100) ## setting grid
kfold = sklearn.model_selection.KFold(10, random_state = 14107, shuffle = True)
ridgeCV = sklearn.linear_model.ElasticNetCV(alphas = lambdas, l1_ratio = 0,
                                            cv = kfold)
ridgeCV.fit(X, y)
```

```python
## optimized value
mse_list = ridgeCV.mse_path_.mean(1)
indx = np.where(mse_list == np.min(mse_list))
min_mse = mse_list[indx]
min_std = ridgeCV.mse_path_.std(1)[indx]
simple_indx = np.min(np.where(mse_list <= min_mse + min_std / np.sqrt(10)))

## plotting
ridgeCV_fig, ax = plt.subplots(figsize=(8,8))
ax.errorbar(-np.log(lambdas), mse_list,
            yerr=ridgeCV.mse_path_.std(1) / np.sqrt(10))
ax.axvline(-np.log(ridgeCV.alpha_), c='k', ls='--')
ax.axhline(min_mse + min_std / np.sqrt(10), c = 'k', ls = '--')
ax.set_ylim([3e10 ,1.6e11])
ax.set_xlabel('$-\log(\lambda)$', fontsize=20)
ax.set_ylabel('Cross-validated MSE', fontsize=20);
ax.scatter(-np.log(lambdas[simple_indx]), mse_list[simple_indx],
           color = "red", s = 15, label = "1-se rule selection", zorder = 5)
plt.legend()
plt.show()

## showing parameter
optim_ridge = lambdas[simple_indx]
print(f'선택된 lambda = {optim_ridge : .5f}')
```

> Ridge에서는 $\lambda = 0.34305$로 선택되었다.

&nbsp;

`-` **Lasso**

```python
## fitting and cross validation(10-fold)
alphas = 10**np.linspace(6, 1.5, 100) ## setting grid for plotting
lassoCV = sklearn.linear_model.ElasticNetCV(alphas = alphas, l1_ratio = 1,
                                            cv = kfold)
lassoCV.fit(X, y)
```

```python
## optimized value
mse_list = lassoCV.mse_path_.mean(1)
indx = np.where(mse_list == np.min(mse_list))
min_mse = mse_list[indx]
min_std = lassoCV.mse_path_.std(1)[indx]
simple_indx = np.min(np.where(mse_list <= min_mse + min_std / np.sqrt(10)))

## plotting
lassoCV_fig, ax = plt.subplots(figsize=(8,8))
ax.errorbar(-np.log(lassoCV.alphas_), lassoCV.mse_path_.mean(1),
            yerr=lassoCV.mse_path_.std(1) / np.sqrt(10))
ax.set_ylim([3e10 ,1.6e11])
ax.axvline(-np.log(lassoCV.alpha_), c='k', ls='--')
ax.axhline(min_mse + min_std / np.sqrt(10), c = 'k', ls = '--')
ax.set_xlabel('$-\log(\lambda)$', fontsize=20)
ax.set_ylabel('Cross-validated MSE', fontsize=20)
ax.scatter(-np.log(lassoCV.alphas_[simple_indx]), mse_list[simple_indx],
           color = "red", s = 15, label = "1-se rule selection", zorder = 5)
plt.legend()
plt.show()

## showing parameter
optim_lasso = lassoCV.alphas_[simple_indx]
print(f'선택된 lambda = {optim_lasso : .5f}')
```

> Lasso에서는 $\lambda = 15199.11083$으로 선택되었다.

#### **문제 2**

2. **해당 $\lambda$로 훈련 데이터 전체에서 $\text{RIDGE}$와 $\text{LASSO}$를 적합하고 평가 데이터에서의 예측오차를 구하시오. (코드 및 결과)**

---

&nbsp; 교차검증을 통해 산출된 최적의 $\lambda$값들을 이용하여 훈련 데이터 전체에서 모형을 적합하고, 평가 데이터에서의 예측오차를 산출하고자 한다.

```python
## Ridge
ridge_predictr = sklearn.linear_model.ElasticNet(alpha = optim_ridge, l1_ratio = 0)
ridge_predictr.fit(X, y)
yyhat_ridge = ridge_predictr.predict(XX) ## prediction

## Lasso
lasso_predictr = sklearn.linear_model.ElasticNet(alpha = optim_lasso, l1_ratio = 1)
lasso_predictr.fit(X, y)
yyhat_lasso = lasso_predictr.predict(XX) ## prediction

## 평가데이터에서의 예측오차 산출
eehat_ridge = yy - yyhat_ridge
eehat_lasso = yy - yyhat_lasso
```

&nbsp; 모형의 적합도를 판단하기 위해 산출된 예측오차에 대하여 평균제곱오차(MSE)와 평균절대오차(MAE)를 계산하면 아래와 같았다.

```python
print(f'Ridge test MSE = {np.mean(eehat_ridge**2):.2f}')
print(f'Ridge test MAE = {np.mean(abs(eehat_ridge)):.2f}')
print(f'Lasso test MSE = {np.mean(eehat_lasso**2):.2f}')
print(f'Lasso test MAE = {np.mean(abs(eehat_lasso)):.2f}')
```

> Ridge Regression의 결과가 더 좋았으므로, 해당 모형을 최종 선택한다.
